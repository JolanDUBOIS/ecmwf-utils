```python
from pathlib import Path
import pandas as pd
```

```python
BASE_DIR = Path("~/Documents/NTNU/Aneo Project")
```

```python
WEATHER_DATA_PATH = BASE_DIR / "data" / "penmanshiel-2017" / "interpolated_data_2017.csv"
PENMANSHIEL_DATA_DIR = BASE_DIR / "data" / "penmanshiel-power"
```

```python
weather_data = pd.read_csv(WEATHER_DATA_PATH)
penmanshiel_static = pd.read_csv(BASE_DIR / "data" / "Penmanshiel_WT_static.csv")
status_data_01 = pd.read_csv(PENMANSHIEL_DATA_DIR / "Penmanshiel_SCADA_2017_WT01-10_3114" / "Status_Penmanshiel_01_2017-01-01_-_2018-01-01_1042.csv", skiprows=9)
turbine_data_01 = pd.read_csv(PENMANSHIEL_DATA_DIR / "Penmanshiel_SCADA_2017_WT01-10_3114" / "Turbine_Data_Penmanshiel_01_2017-01-01_-_2018-01-01_1042.csv", skiprows=9)
```

```python
# Rename issued col by issued_time
weather_data = weather_data.rename(columns={"issued": "issued_time", "time": "forecast_time"})
weather_data.head()
```

```python
weather_data["issued_time"] = pd.to_datetime(weather_data["issued_time"], errors="coerce")
weather_data["forecast_time"] = pd.to_datetime(weather_data["forecast_time"], errors="coerce")
status_data_01["Timestamp start"] = pd.to_datetime(status_data_01["Timestamp start"], errors="coerce")
status_data_01["Timestamp end"] = pd.to_datetime(status_data_01["Timestamp end"], errors="coerce")
turbine_data_01["Date and time"] = pd.to_datetime(turbine_data_01["# Date and time"], errors="coerce")
turbine_data_01.drop(columns=["# Date and time"], inplace=True)
```

```python
weather_data.head()
```

```python
static_01 = penmanshiel_static[penmanshiel_static["Title"] == "Penmanshiel 01"].copy()
static_01
```

```python
status_data_01.drop_duplicates(inplace=True)
status_data_01.head()
```

The status tables are very interesting and could be used but we would need to discuss HOW to use them because there is a need for a lot of parsing of the different columns to get a clean "stopped"/"running" kind of column.

```python
# This can easily be changed to include more or less columns

COLS_TO_KEEP = [
    "Date and time",
    "Energy Export (kWh)",
    # "Energy Export counter (kWh)",
    "Energy Import (kWh)",
    # "Energy Import counter (kWh)",
    "Power (kW)",
    # "Power, Standard deviation (kW)",
    # "Power, Minimum (kW)",
    # "Power, Maximum (kW)"
]

turbine_data_01_reduced = turbine_data_01[COLS_TO_KEEP]
turbine_data_01_reduced.head()
```

```python
turbine_data_01_reduced.columns
```

```python
# Group turbine_data_01_reduced by hour, taking the sum of Energy Export and Energy Import, and the mean of Power

turbine_data_01_reduced.set_index("Date and time", inplace=True)
turbine_01_hourly = turbine_data_01_reduced.resample("h").agg({
    "Energy Export (kWh)": "sum",
    "Energy Import (kWh)": "sum",
    "Power (kW)": "mean"
}).reset_index()

turbine_01_hourly.head()
```

```python
print(f"The time range of status_data_01 is from {status_data_01['Timestamp start'].min()} to {status_data_01['Timestamp end'].max()}")
print(f"The time range of turbine_data_01 is from {turbine_data_01['Date and time'].min()} to {turbine_data_01['Date and time'].max()}")
print(f"The time range of weather_data is from {weather_data['forecast_time'].min()} to {weather_data['forecast_time'].max()}")
```

```python
start = pd.to_datetime("2017-01-01 00:00:00")
end = pd.to_datetime("2017-12-31 23:50:00")
```

```python
weather_data_filtered = weather_data[(weather_data["forecast_time"] >= start) & (weather_data["forecast_time"] <= end)]
weather_data_filtered = weather_data_filtered[(weather_data_filtered["latitude"] == static_01["Latitude"].values[0]) & (weather_data_filtered["longitude"] == static_01["Longitude"].values[0])]
```

```python
weather_data_filtered.head(30)
```

```python
df = weather_data_filtered.merge(
    turbine_01_hourly,
    left_on="forecast_time",
    right_on="Date and time",
    how="left"
)
df.drop(columns=["Date and time"], inplace=True)
```

```python
df.columns
```

```python
df['temperature_K'] = df['t2m']
df['temperature_C'] = df['temperature_K'] - 273.15
df.drop(columns=['t2m'], inplace=True)
```

```python
cols_to_show = [
    'issued_time', 'forecast_time',
    'u10', 'v10', 'msl', 'tp', 'd2m',
    'temperature_C',
    'temperature_K', 'Energy Export (kWh)',
    'Energy Import (kWh)',
    'Power (kW)'
]
df[cols_to_show].head()
```
